```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import pandas as pd
from pandas import ExcelFile

sns.set(style="whitegrid")
```

```python
df_renda = pd.read_excel('ipea/ipeadata[30-08-2019-02-34]_rendamedia.xls', 'Séries')
df_renda.head()
```

```python
sns.set()

df = df_renda.sort_values(by=['Região'])

plt.figure(figsize=(16, 10))
fig = sns.barplot(x="Região", y="2014", data=df)

plt.title('Renda Média da População das Regiões Brasileiras em 2014', fontsize=20, position=[.5, 1.05])

plt.xlabel('Região', fontsize=16)
plt.ylabel('Valor médio em Reais(R$)', fontsize=16)

plt.annotate('Fonte: Portal do IPEA (http://www.ipea.gov.br)', (0, 0), (650, -60), xycoords='axes fraction', textcoords='offset points') 
```

# Gráficos de Barras Empilhadas

```python
df_populacao = pd.read_excel('ipea/ipeadata[30-08-2019-02-45]populacao.xls', 'Séries')
df_populacao.head()
```

```python
df = df_populacao

sns.set()
df2 = df.drop(['Sigla', 'Codigo'], axis=1)
```

```python
sns.set(style="whitegrid")

fig = df2.set_index('Região').T.plot(kind='bar', stacked=True, figsize=(16, 10))
plt.title('Contagem da População Brasileira por Regiões', fontsize=20, position=[.5, 1.05])

mkfunc = lambda x, pos: '%1.1fM' % (x * 1e-6) if x >= 1e6 else '%1.1fK' % (x * 1e-3) if x >= 1e3 else '%1.1f' % x
mkformatter = ticker.FuncFormatter(mkfunc)
fig.yaxis.set_major_formatter(mkformatter)

plt.setp(fig.get_xticklabels(), rotation=0)

plt.xlabel('Ano', fontsize=16)
plt.ylabel('Número de Indivíduos', fontsize=16)

plt.annotate('Fonte: Portal do IPEA (http://www.ipea.gov.br)', (0, 0), (650, -60), xycoords='axes fraction', textcoords='offset points')
```

```python
sns.set(style="whitegrid")

df2 = df2.sort_values(by=['1991'], ascending=False)
fig = df2.set_index('Região').T.plot(kind='barh', stacked=True, figsize=(16, 10))
plt.title('Contagem da População Brasileira por Regiões', fontsize=20, position=[.5, 1.05])

plt.ylabel('Ano', fontsize=16)
plt.xlabel('Número de Indivíduos', fontsize=16)


mkfunc = lambda x, pos: '%1.1fM' % (x * 1e-6) if x >= 1e6 else '%1.1fK' % (x * 1e-3) if x >= 1e3 else '%1.1f' % x
mkformatter = ticker.FuncFormatter(mkfunc)
fig.xaxis.set_major_formatter(mkformatter)

plt.annotate('Fonte: Portal do IPEA (http://www.ipea.gov.br)', (0, 0), (650, -60), xycoords='axes fraction', textcoords='offset points')
```

```python
df3 = df2.copy()
#df3.loc[:,'Total'] = df3.sum(axis=1)
df3
df3.iloc[:, 1:6] = df3.iloc[:, 1:6].apply(lambda x: x / x.sum(), axis=0)
df3
```

```python
sns.set(style="whitegrid")
df3 = df3.sort_values(by=['1991'], ascending=False)
fig = df3.set_index('Região').T.plot(kind='barh', stacked=True, figsize=(16, 10))
plt.title('Proporção da População Brasileira por Regiões', fontsize=20, position=[.5, 1.05])

plt.ylabel('Ano', fontsize=16)
plt.xlabel('Porcentual da População', fontsize=16)

fig.xaxis.set_major_formatter(ticker.PercentFormatter(xmax=1.0))

plt.annotate('Fonte: Portal do IPEA (http://www.ipea.gov.br)', (0, 0), (650, -60), xycoords='axes fraction', textcoords='offset points')
```
