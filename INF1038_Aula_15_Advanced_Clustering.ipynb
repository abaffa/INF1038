## Spectral Clustering

#### Dataset Waves

```python
import numpy as np

from sklearn.preprocessing import StandardScaler

nb_samples = 1000

wave_X = np.zeros(shape=(nb_samples, 2))

for i in range(nb_samples):
    wave_X[i, 0] = float(i)
    
    if i % 2 == 0:
        wave_X[i, 1] = 1.0 + (np.random.uniform(0.65, 1.0) * np.sin(float(i) / 100.0))
    else:
        wave_X[i, 1] = 0.1 + (np.random.uniform(0.5, 0.85) * np.sin(float(i) / 100.0))
        
ss = StandardScaler()
wave_Xs = ss.fit_transform(wave_X)
```

```python
import matplotlib.pyplot as plt
plt.viridis()
plt.scatter(wave_X[:, 0], wave_X[:, 1])
plt.show()
```

#### Spectral Clustering n=2, affinity=nn

```python
from sklearn.cluster import SpectralClustering

sc = SpectralClustering(n_clusters=2, affinity='nearest_neighbors', n_neighbors=20, random_state=1000)
wave_Y_sc = sc.fit_predict(wave_Xs)
```

```python
import matplotlib.pyplot as plt
plt.viridis()
plt.scatter(wave_X[:, 0], wave_X[:, 1],c=wave_Y_sc)
plt.show()
```

#### Comparação com k-means, k=2

```python
from sklearn.cluster import KMeans

km = KMeans(n_clusters=2, random_state=1000)
Y_km = km.fit_predict(wave_Xs)
```

```python
import matplotlib.pyplot as plt
plt.viridis()
plt.scatter(wave_X[:, 0], wave_X[:, 1],c=km.predict(wave_Xs))
plt.show()
```

#### Comparação com DBScan

```python
epss = [1, 5]
i=0

colors = np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta"])

for eps in epss:
    i = i + 1
    dbs = DBSCAN(eps=eps)
    #plt.scatter(wave_X[:, 0], wave_X[:, 1],c=colors[dbs.fit_predict(wave_X)])
    plt.scatter(wave_X[:, 0], wave_X[:, 1],c=dbs.fit_predict(wave_X))
    plt.show()
```

#### Spectral Clustering n=2, affinity=RBF

```python
import numpy as np

from sklearn.cluster import SpectralClustering

gammas = np.linspace(0, 12, 4)
 
for gamma in gammas:
    sc = SpectralClustering(n_clusters=2, affinity='rbf', gamma=gamma)
    plt.scatter(wave_X[:, 0], wave_X[:, 1],c=sc.fit_predict(wave_X))
    plt.show()
```

### Connectivity constraints

```python
from sklearn.datasets import make_circles

nb_samples = 3000
X, Y = make_circles(n_samples=nb_samples, noise=0.05, random_state=1000)
```

```python
import matplotlib.pyplot as plt
plt.viridis()
plt.scatter(X[:, 0], X[:, 1])
plt.show()
```

```python
from sklearn.metrics import silhouette_score

ac = AgglomerativeClustering(n_clusters=20, linkage='average')
Y_pred = ac.fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c=Y_pred)
plt.show()
    
print('Silhouette score: %.3f' % silhouette_score(X, Y_pred))
```

```python
from sklearn.neighbors import kneighbors_graph
from sklearn.metrics import silhouette_score

acc = []
k = [50, 100, 200, 500]

for i in k:
    kng = kneighbors_graph(X, i)
    ac1 = AgglomerativeClustering(n_clusters=20, connectivity=kng, linkage='average')
    Y_pred = ac1.fit_predict(X)
    print('Silhouette score (k=%d): %.3f' % (i, silhouette_score(X, Y_pred)))
    plt.scatter(X[:, 0], X[:, 1], c=Y_pred)
    plt.show()
    acc.append(ac1)
```

## BIRCH

```python
from sklearn.datasets import make_blobs

nb_samples = 2000

X, Y = make_blobs(n_samples=nb_samples, n_features=2, centers=5, cluster_std=1.5, random_state=1000)
```

```python
from sklearn.cluster import Birch
from sklearn.metrics import adjusted_rand_score

birch = Birch(n_clusters=5, threshold=0.15, branching_factor=100)

batch_size = 80

X_batch = []
Y_preds = []

for i in range(0, nb_samples, batch_size):
    birch.partial_fit(X[i:i + batch_size])
    X_batch.append(X[:i + batch_size])
    Y_preds.append(birch.predict(X[:i + batch_size])) 

print(adjusted_rand_score(birch.predict(X), Y))
```

```python
for i in range(0,len(Y_preds)):
    
    X = X_batch[i]
    Y = Y_preds[i]
    print(len(X),"amostras")
    plt.scatter(X[:, 0], X[:, 1], c=Y)
    plt.show()
```

## Biclustering

```python
import numpy as np

nb_users = 100
nb_products = 150
max_rating = 10

np.random.seed(1000)

up_matrix = np.random.randint(0, max_rating + 1, size=(nb_users, nb_products))
mask_matrix = np.random.randint(0, 2, size=(nb_users, nb_products))
up_matrix *= mask_matrix
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.imshow(up_matrix, cmap='hot', interpolation='nearest')
plt.xlabel('Produtos')
plt.ylabel('Usuários')
plt.colorbar()
plt.show()
```

```python
import numpy as np
from sklearn.cluster import SpectralBiclustering

k = 2

sbc = SpectralBiclustering(n_clusters=k, random_state=1000)
sbc.fit(up_matrix)

up_clustered = np.outer(np.sort(sbc.row_labels_) + 1, np.sort(sbc.column_labels_) + 1)
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.imshow(up_clustered, cmap='hot', interpolation='nearest')
plt.xlabel('Produtos')
plt.ylabel('Usuários')
plt.xticks([])  
plt.yticks([])  
plt.show()
```

```python
for i in range(0,k*k):
    print("Usuários no cluster",i)
    print(np.where(sbc.rows_[i, :] == True))
```

```python
for i in range(0,k*k):
    print("Usuários no cluster",i)
    print(np.where(sbc.columns_[i, :] == True))
```
