### Multinomial Naive Bayes

```python
#http://qwone.com/~jason/20Newsgroups/

from sklearn.datasets import fetch_20newsgroups_vectorized

train_data = fetch_20newsgroups_vectorized(subset='train')
test_data = fetch_20newsgroups_vectorized(subset='test')
```

```python
from sklearn.naive_bayes import MultinomialNB

mnb = MultinomialNB(alpha=0.01)
mnb.fit(train_data['data'], train_data['target'])

print(mnb.score(test_data['data'], test_data['target']))
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.metrics import confusion_matrix

pred_y = mnb.predict(test_data['data'])
cm = confusion_matrix(pred_y, test_data['target'])

disp = ConfusionMatrixDisplay(confusion_matrix=cm,  display_labels=test_data.target_names)
fig, ax = plt.subplots(figsize=(10,10))
disp.plot(ax=ax,xticks_rotation = 'vertical')
plt.show()
```

```python
from sklearn.preprocessing import binarize

X_train_binary = binarize(train_data['data'], threshold=0.01)
X_test_binary = binarize(test_data['data'], threshold=0.01)
```

### Gaussian Naive Bayes

```python
from sklearn.datasets import make_classification

nb_samples = 300

X, Y = make_classification(n_samples=nb_samples, n_features=2, n_informative=2, n_redundant=0)
```

```python
import matplotlib.pyplot as plt
plt.viridis()
plt.scatter(X[:, 0], X[:, 1], c=Y)
plt.show()
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_curve, auc
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25)

gnb = GaussianNB()
gnb.fit(X_train, Y_train)
Y_gnb_score = gnb.predict_proba(X_test)

lr = LogisticRegression()
lr.fit(X_train, Y_train)
Y_lr_score = lr.decision_function(X_test)

fpr_gnb, tpr_gnb, thresholds_gnb = roc_curve(Y_test, Y_gnb_score[:, 1])
fpr_lr, tpr_lr, thresholds_lr = roc_curve(Y_test, Y_lr_score)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 8))
plt.plot(fpr_gnb, tpr_gnb, color='red', label='GaussianNB (AUC: %.2f)' % auc(fpr_gnb, tpr_gnb))
plt.plot(fpr_lr, tpr_lr, color='green', label='LogisticRegression (AUC: %.2f)' % auc(fpr_lr, tpr_lr))
plt.plot([0, 1], [0, 1], color='blue', linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.01])
plt.title('Curva ROC')
plt.xlabel('Taxa de Falsos Positivos')
plt.ylabel('Taxa de Verdadeiros Positivos')
plt.legend(loc="lower right")

plt.show()
```

```python
from sklearn.datasets import load_digits
from sklearn.model_selection import cross_val_score

digits = load_digits()

gnb = GaussianNB()
mnb = MultinomialNB()
lr = LogisticRegression(max_iter=5000)

cv_score_gnb = cross_val_score(gnb, digits.data, digits.target, scoring='accuracy', cv=10).mean()
print(cv_score_gnb)

cv_score_mnb = cross_val_score(mnb, digits.data, digits.target, scoring='accuracy', cv=10).mean()
print(cv_score_mnb)

cv_score_lr = cross_val_score(lr, digits.data, digits.target, scoring='accuracy', cv=10).mean()
print(cv_score_lr)
```

### Discriminant analysis

```python
from sklearn.datasets import make_blobs

nb_samples = 1000

X, Y = make_blobs(n_samples=nb_samples, n_features=2, centers=2, cluster_std=[1.0, 10.0], random_state=1000)
```

```python
import matplotlib.pyplot as plt
plt.viridis()
plt.scatter(X[:, 0], X[:, 1], c=Y)
plt.show()
```

```python
import numpy as np

print(np.cov(X[Y==0].T))
print(np.cov(X[Y==1].T))
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis, QuadraticDiscriminantAnalysis

lda = LinearDiscriminantAnalysis()
print(cross_val_score(lda, X, Y, cv=10).mean())

qda = QuadraticDiscriminantAnalysis()
print(cross_val_score(qda, X, Y, cv=10).mean())
```
